````python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
````

````python
df_train = pd.read_csv("C:\\Users\\xpelu\\OneDrive\\Escritorio\\titanic\\train.csv")
print(df_train.shape)
#df_train = df_train.dropna() #usando solo esto se pierden muchos datos de entrenamiento
print(df_train.shape)
df_train.head()
````

````python
df_test = pd.read_csv("C:\\Users\\xpelu\\OneDrive\\Escritorio\\titanic\\test.csv")
df_test.head()
````

````python
test_nan = []
for i in df_test.columns.tolist():
    flag = True
    for j in df_test[i].unique().tolist():
        if pd.isnull(j):
            flag = False
    if not flag:
        test_nan.append(i)
for i in test_nan:
    cont_total = 0
    cont_nan = 0
    for j in df_test[i]:
        cont_total += 1
        if pd.isnull(j):
            cont_nan += 1
    print("porcentaje de nan en "+ i +": ", cont_nan/cont_total*100, "%")
````

Lo anterior es para ver que columnas del conjunto de testeo tienen datos faltantes. Con esto se tiene que se puede descartar la columna "Cabin" para entrenar ya que para el conjunto de testeo no va a tener un aporte a la hora de predecir el valor objetivo al tener tantos datos faltantes.

````python
train_nan = []
for i in df_train.columns.tolist():
    flag = True
    for j in df_train[i].unique().tolist():
        if pd.isnull(j):
            flag = False
    if not flag:
        train_nan.append(i)
for i in train_nan:
    cont_total = 0
    cont_nan = 0
    for j in df_train[i]:
        cont_total += 1
        if pd.isnull(j):
            cont_nan += 1
    print("porcentaje de nan en "+ i +": ", cont_nan/cont_total*100, "%")
````

Con estos resultados se haran 2 pruebas: La primera entrenando considerando la edad y la segunda sin entrenar la edad (cambiar si no me da el tiempo xd)

````python
df_train_aux = df_train.copy()
df_train_aux = df_train_aux.drop("Cabin", axis= 1)
print(df_train_aux.shape)
df_train_aux = df_train_aux.dropna()
print(df_train_aux.shape)
df_train_aux.head() 
````

Ahora la distribucion de las variables continuas

````python
variables_cont = ["Age", "Fare"]
for i in variables_cont:
    sns.boxplot(x=i, data=df_train_aux)
    plt.show()
````

Se tienen muchos outliers por lo que se va a considerar a la hora de normalizar. Ahora a ver la distribucion de las variables categoricas que se consideraran.

````python
variables_cat = ["Pclass", "Sex", "SibSp", "Parch", "Embarked"]

for i in variables_cat:
    sns.countplot(x=i, data=df_train_aux)
    plt.show()
````

````python
#embarked y sex
orden_sex = {'male': 1, 'female': 2}
orden_embarked = {"S": 1, "C": 2, "Q": 3}
orden = []
orden.append(orden_sex)
orden.append(orden_embarked)
cont_aux = 0
for i in ["Sex", "Embarked"]:
    df_train_aux[i] = df_train_aux[i].map(orden[cont_aux])
    cont_aux += 1
df_train_aux.head()
````

Ahora a normalizar las variables continuas.

````python
from sklearn.preprocessing import MinMaxScaler

aux = []
for i in variables_cont:
    cont = 0
    for j in df_train_aux.columns.tolist():
        if i == j:
            aux.append(cont)
        cont += 1

scaler = MinMaxScaler()
X = np.array(df_train_aux.iloc[:, aux])
scaler = MinMaxScaler().fit(X)
X_scaler = scaler.transform(X)
print(X_scaler.shape)
print(X_scaler)
````

````python
feature_arr = df_train_aux[variables_cat]
final_arr = np.concatenate((feature_arr, X_scaler),axis=1)
print(final_arr.shape)
print(final_arr)
````

Ahora el y objetivo.

````python
y_train = np.array(df_train_aux["Survived"])
print(y_train.shape)
````

````python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(final_arr, y_train)
````

````python
y_pred = model.predict(final_arr)
````

Prueba del modelo para el mismo conjunto de entrenamiento.

````python
from sklearn.metrics import classification_report
print(classification_report(y_train, y_pred))
````
