### 4- Preprocesamiento

```python
df_train["Transported"] = df_train["Transported"].astype(int)
df_train.head()
```

```python
print(df_train.shape)
df_train = df_train.dropna() #esto bota casi el 25% de los datos
print(df_train.shape)
```

```python
from sklearn.preprocessing import RobustScaler

aux = []
for i in continuas:
    cont = 0
    for j in df_train.columns.tolist():
        if i == j:
            aux.append(cont)
        cont += 1

X = np.array(df_train.iloc[:, aux])
scaler = RobustScaler().fit(X)
X_scaler = scaler.transform(X)
print(X_scaler.shape)
print(X_scaler)
```

```python
from sklearn.preprocessing import LabelEncoder

for i in categoricas:
    df_train[i] = LabelEncoder().fit_transform(df_train[i])
```

```python
X = df_train[categoricas]
final_arr = np.concatenate((X_scaler, X),axis=1)
data_aux = pd.DataFrame(final_arr, columns = continuas + categoricas)
```

```python
y = df_train["Transported"].tolist()
```

### 5- Seleccion de modelo

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import classification_report

model1 = Perceptron().fit(final_arr,y)
y_pred1 = model1.predict(final_arr)
print(classification_report(y_pred1, y))
```

```python
from sklearn.ensemble import RandomForestClassifier
estimadores = [1,10,100,1000]
for i in estimadores:
    model2 = RandomForestClassifier(n_estimators=i).fit(final_arr,y)
    y_pred2 = model2.predict(final_arr)
    print(classification_report(y_pred2, y))
#despues de 100 la precision no mejora
```

```python
from sklearn.svm import SVC

C_values = [0.1,1,10,100,1000,10000]

for i in C_values:
    model3 = SVC(C=i).fit(final_arr, y)
    y_pred3 = model3.predict(final_arr)
    print(classification_report(y_pred3, y))
#despues de 1000 mejora pero no tanto y el tiempo de ejecucion sube mucho
```

```python
from sklearn.neighbors import KNeighborsClassifier

neighbors = [1,2,3,4] #primero se probo con [1,10,100,1000], pero la precision era demasiado mala asi que se tomaron valores mas bajos
for i in neighbors:
    model4 = KNeighborsClassifier(n_neighbors=i).fit(final_arr,y)
    y_pred4 = model4.predict(final_arr)
    print(classification_report(y_pred4, y))
#para n_neighbors = 1 se tiene la mejor precision
```

### 6- Metricas y analisis de resultados

* Las metricas estan en el punto 5 (con los classification_report). 
* De los 4 modelos anteriores se tiene que el modelo que mejor predice es RandomForestClassifier para n_estimators = 100.
* El que peor predijo fue el percerptron.
