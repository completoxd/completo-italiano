Aqui se juntan los 2 dataframes que tienen una columna (lp_id) en comun para juntar la mayor parte de informacion posible para luego ver el comportamiento de los datos para algunas fechas.

```python
df_2_3 = pd.merge(df3, df2, on="lp_id", how="inner")
print(df_2_3.shape)
df_2_3.head()
```

```python
df_2_3 = df_2_3.dropna()
print(df_2_3.shape)
```

```python
df_aux = df_2_3.sort_values(by="time").groupby(by=["time"])
df_aux.head()
```

Aqui se seleccionan algunas fechas para ver ciertos valores de una muestra para ver que es lo que pasa con respecto a las variables usando describe().

```python
lista_aux = []
for i in range(len(df_2_3["time"].unique())):
    if i%30 == 0:
        lista_aux.append(df_2_3["time"].unique()[i])
for j in lista_aux:
    aux = df_2_3[df_2_3["time"]==j]
    print(j +": \n", aux.describe())
```

```python
for j in lista_aux:
    aux = df_2_3[df_2_3["time"]==j]
    print(j +": \n")
    sns.countplot(x="Primary Essential Function 1", data=aux)
    plt.show()
```
